```python
# remove outliers
from scipy import stats
# calculate the z-score for each value in the DataFrame
z_scores = stats.zscore(df)
# create a boolean mask for values that are not outliers
mask = (z_scores < 3).all(axis=1)
# apply the mask to the DataFrame to remove outliers
df = df[mask]
```

```python
# Select only the best 5 features to predict, i selected the remaining features becuase of the feature importance and low correlation between features
df=df.drop(['Current','Frequency','WORKSTATION_GPU','WORKSTATION_GPU_TEMPRETURE','WORKSTATION_GPU_POWER_CONSUMPTION','WORKSTATION_CPU_POWER_consumption','WORKSTATION_RAM_POWER_CONSUMPTION','WORKSTATION_RAM','Weekday','WORKSTATION_CPU_TEMPRETURE'],axis=1)
df.head()
```

```python
#Defining X and Y
Y=df['Active energy']
X=df.drop(['Active energy'],axis=1)
print("X columns are: ",X.columns)
print("number of X columns : ", len(X.columns))
```

```python
# Split into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# **Minute**

## **Random Forest**

```python
#Build Random forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
rf_regressor = RandomForestRegressor(random_state=42)

# Define the best hyperparameter
param_dist = {
    'n_estimators': randint(100, 300),
    'max_depth': [5, 10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform a randomized search over the parameter distribution
random_search = RandomizedSearchCV(rf_regressor, param_distributions=param_dist, n_iter=10, cv=5, random_state=42)

# Fit the randomized search to the data
random_search.fit(X_train, Y_train)

# Print the best hyperparameters
print('Best hyperparameters:', random_search.best_params_)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_regressor = RandomForestRegressor(n_estimators=249,max_depth=20,random_state=42,min_samples_split=10,min_samples_leaf=1)
rf_regressor.fit(X_train, Y_train)
```

```python
# Use the trained model to make predictions on the test set
y_pred = rf_regressor.predict(X_test)
y_pred
```

```python
# Evaluate the performance of the model using Mean Squared Error and R2 Score metrics
from sklearn.metrics import mean_squared_error
import math
mse = mean_squared_error(Y_test, y_pred)
print("Mean Squared Error:", mse)
# Calculate RMSE
rmse = math.sqrt(mse)
print("RMSE value:", rmse)
```

```python
#Feature importance
import matplotlib.pyplot as plt
importances = rf_regressor.feature_importances_
# get the names of the features
feature_names = X.columns
# sort the features by their importances
indices = np.argsort(importances)
# plot the feature importances
plt.figure(figsize=(10, 8))
plt.barh(range(len(indices)), importances[indices])
plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
plt.xlabel('Feature Importance')
plt.show()
```
